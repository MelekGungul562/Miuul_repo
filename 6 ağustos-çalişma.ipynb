```python
# Rastgele bir NumPy array oluÅŸturalÄ±m
np.random.seed(19)  # AynÄ± sonuÃ§larÄ± almak iÃ§in
# data = np.append(np.random.normal(50, 10, 100), [5, 120])  # Outlier ekleyelim
# data = np.append(np.random.randint(40, 60, 100), [5, 120])
data = np.append(np.round(np.random.normal(50, 10, 100)), [5, 10, 20, 80, 90, 110, 120]).astype(int)
data
```

```python
# Histogram Ã§iz
plt.hist(data, bins=50, color='skyblue', edgecolor='black')
mean_value = np.mean(data)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Ortalama: {mean_value:.2f}')
plt.title("Histogram GrafiÄŸi")
plt.xlabel("DeÄŸerler")
plt.ylabel("Frekans")
plt.show();
```

```python
# Histogram Ã§iz
plt.boxplot(data, vert=False)
plt.title("Boxplot GrafiÄŸi")
plt.xlabel("Value")
plt.ylabel("Feature")
plt.show();
```

```python
# IQR hesaplama
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
# Alt ve Ã¼st sÄ±nÄ±rlarÄ± belirleme
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("lower_bound :", lower_bound)
print("upper_bound :", upper_bound)
```

```python
# numpy.where() koÅŸulun doÄŸru olduÄŸu Ã¶ÄŸelerin indekslerini dÃ¶ndÃ¼rÃ¼r.
outlier_index = np.where((data < lower_bound) | (data > upper_bound))[0]
```

```python
type(np.where((data < lower_bound) | (data > upper_bound))[0])
```
